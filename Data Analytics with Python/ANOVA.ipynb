```python
import pandas as pd
import numpy as np
import math
from scipy import stats
import scipy
import statsmodels.api as sm
from statsmodels.formula.api import ols
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel('C:/Users/shubham.chhetri/Desktop/MYPROJECT/Python/oneway.xlsx')
```

```python
newdf=pd.melt(df.reset_index(),id_vars=['index'],value_vars=['tec1','tec2','tec3'])
newdf
```

```python
newdf.columns=['index','treat','value']
```

```python
newdf
```

```python
model=ols('value ~ C(treat)',data=newdf).fit()
```

```python
anova_tb=sm.stats.anova_lm(model,typ=1)
anova_tb
```
